## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns=['EIN','NAME'])
application_df
```

```python
# Determine the number of unique values in each column.
unique_counts = application_df.nunique()
unique_counts
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
application_counts = application_df['APPLICATION_TYPE'].value_counts()
print(application_counts)
```

```python
# Choose a cutoff value and create a list of application types to be replaced
threshold = 200
apps_to_replace = application_df['APPLICATION_TYPE'].value_counts()[application_df['APPLICATION_TYPE'].value_counts() < threshold].index.tolist()

# Replace in dataframe
for app in apps_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app, "Other")

# Check to make sure replacement was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
classification_counts = application_df['CLASSIFICATION'].value_counts()
print(classification_counts)
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classifications_more_than_one = classification_counts[classification_counts > 1]
classifications_more_than_one
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
cutoff = 1000
classifications_to_replace = application_df['CLASSIFICATION'].value_counts()[application_df['CLASSIFICATION'].value_counts() <= cutoff].index.tolist()

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure replacement was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_df_dummies = pd.get_dummies(application_df)
application_df_dummies
```

```python
#turn booleans into binary

booleans = application_df_dummies.select_dtypes(include=['bool']).columns
application_df_dummies[booleans] = application_df_dummies[booleans].astype('int64')
application_df_dummies
```

```python
# Split our preprocessed data into our features and target arrays
y = application_df_dummies['IS_SUCCESSFUL']  # Target variable
X = application_df_dummies.drop('IS_SUCCESSFUL', axis=1)  # Features

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)  
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

USE KERAS TUNER

```python
# Create a method that creates a new Sequential model with hyperparameter options
def make_model(hp):
    nn = tf.keras.models.Sequential()

    input_dim = len(X_train_scaled[0])

    # Allow kerastuner to decide which activation function to use in hidden layers
    activate = hp.Choice('activation',['relu','tanh'])

    # Allow kerastuner to decide number of neurons in first layer
    nn.add(tf.keras.layers.Dense(units=hp.Int('first_units',
        min_value=1,
        max_value=80,
        step=5), activation=activate, input_dim=input_dim))

    # Allow kerastuner to decide number of hidden layers and neurons in hidden layers
    for i in range(hp.Int('num_layers', 1, 5)):
        nn.add(tf.keras.layers.Dense(units=hp.Int('units_' + str(i),
            min_value=1,
            max_value=40,
            step=5),
            activation=activate))

    nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

    # Compile the model
    nn.compile(loss="binary_crossentropy", optimizer='adam', metrics=["accuracy"])

    return nn
```

```python
# Import the kerastuner library
import keras_tuner as kt

tuner = kt.Hyperband(
    make_model,
    objective="val_accuracy",
    max_epochs=20,
    hyperband_iterations=2)
```

```python
tuner.search(X_train_scaled,y_train,epochs=20,validation_data=(X_test_scaled,y_test))
```

```python
# Evaluate the top 3 models against the test dataset
top_model = tuner.get_best_models(3)
for model in top_model:
    model_loss, model_accuracy = model.evaluate(X_test_scaled,y_test,verbose=2)
    print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
top_model_hyper = tuner.get_best_hyperparameters(1)[0]
top_model_hyper.values
```

## Compile, Train and Evaluate the Model using the best Keras Tuner models

```python


# Initialize the Model
model = tf.keras.models.Sequential()

# Add the first hidden layer with specified units and activation function
model.add(tf.keras.layers.Dense(units=71, activation='relu', input_dim=43))  # Assuming input_dim=2 for a 2D input

# Add additional hidden layers based on the specified units
model.add(tf.keras.layers.Dense(units=11, activation='relu'))  # units_0
model.add(tf.keras.layers.Dense(units=16, activation='relu'))  # units_1
model.add(tf.keras.layers.Dense(units=6, activation='relu'))   # units_2
model.add(tf.keras.layers.Dense(units=11, activation='relu'))  # units_3
model.add(tf.keras.layers.Dense(units=1, activation='sigmoid')) # units_4 (output layer for binary classification)


# Display the model's architecture
model.summary()
```

```python
# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model
# Fit the model to the training data
model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_test, y_test))
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
model_filename = 'AlphabetSoupCharity_Optimized2.h5'  # Specify the filename for the model
model.save(model_filename)  # Save the model
```
