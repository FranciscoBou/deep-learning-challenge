## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns=['EIN','NAME'])
application_df
```

```python
# Determine the number of unique values in each column.
unique_counts = application_df.nunique()
unique_counts
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
application_counts = application_df['APPLICATION_TYPE'].value_counts()
print(application_counts)
```

```python
# Choose a cutoff value and create a list of application types to be replaced
threshold = 200
apps_to_replace = application_df['APPLICATION_TYPE'].value_counts()[application_df['APPLICATION_TYPE'].value_counts() < threshold].index.tolist()

# Replace in dataframe
for app in apps_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app, "Other")

# Check to make sure replacement was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
classification_counts = application_df['CLASSIFICATION'].value_counts()
print(classification_counts)
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classifications_more_than_one = classification_counts[classification_counts > 1]
classifications_more_than_one
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
cutoff = 1000
classifications_to_replace = application_df['CLASSIFICATION'].value_counts()[application_df['CLASSIFICATION'].value_counts() <= cutoff].index.tolist()

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure replacement was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_df_dummies = pd.get_dummies(application_df)
application_df_dummies
```

```python
#turn booleans into binary

booleans = application_df_dummies.select_dtypes(include=['bool']).columns
application_df_dummies[booleans] = application_df_dummies[booleans].astype('int64')
application_df_dummies

```

```python
# Split our preprocessed data into our features and target arrays
y = application_df_dummies['IS_SUCCESSFUL']  # Target variable
X = application_df_dummies.drop('IS_SUCCESSFUL', axis=1)  # Features

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)  

y_train = y_train.values.reshape(-1, 1)  # Reshape to (n_samples, 1)
y_test = y_test.values.reshape(-1, 1)    # Reshape to (n_samples, 1)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model using the best Keras Tuner models and now optimizing using gradient ascent


```python

# Define your model
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(71, activation='relu', input_dim=X_train.shape[1]),  # Use the number of features in X_train
    tf.keras.layers.Dense(11, activation='relu'),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(6, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')  # Output layer for binary classification
])
```

```python
# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Gradient Ascent parameters
learning_rate = 0.01
epochs = 100
```

```python
# Create an optimizer
optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)
```

```python
# Custom training loop for Gradient Ascent
for epoch in range(epochs):
    with tf.GradientTape() as tape:
        # Forward pass
        predictions = model(X_train, training=True)
        # Compute loss
        loss = tf.keras.losses.binary_crossentropy(y_train, predictions)
        # Compute accuracy
        accuracy = tf.keras.metrics.binary_accuracy(y_train, predictions)

    # Compute gradients (negative because we want to maximize)
    gradients = tape.gradient(loss, model.trainable_variables)
    
    # Apply gradients using the optimizer
    optimizer.apply_gradients(zip(gradients, model.trainable_variables))

    # Print loss and accuracy
    print(f'Epoch {epoch + 1}, Loss: {tf.reduce_mean(loss).numpy()}, Accuracy: {tf.reduce_mean(accuracy).numpy()}')

# After training, you can evaluate the model
model.evaluate(X_test, y_test)
```

```python
# Train the model
# Fit the model to the training data
model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_test, y_test))
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
model_filename = 'AlphabetSoupCharity_Optimized2.h5'  # Specify the filename for the model
model.save(model_filename)  # Save the model
```
